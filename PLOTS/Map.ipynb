```python
import pygmt
```

```python
import pandas as pd
data=pd.read_csv("samples.txt", sep="\t")

internal_data = data[data["sample"] == "internal"]
external_data = data[data["sample"] == "external"]
```

```python
fig = pygmt.Figure()
fig.coast(
    region=[-25,85,20,70],
    resolution="h",
    projection="M0/0/12c",
    land="#f5f5ef",
    water="#b3c1c9",
    shorelines="1/0.2p,grey",
    borders="1/0.2p,grey")

fig.plot(x=internal_data.longtitude, 
         y=internal_data.latitude, 
         style="cc", 
         fill="#e3aa4d", 
         pen="black",
         size=round(internal_data.number*0.01,1)+0.2
        )

fig.plot(x=external_data.longtitude, 
         y=external_data.latitude, 
         style="cc", 
         fill="#41534f", 
         pen="black",
         size=round(external_data.number*0.01,1)+0.2
        )
fig.show()
fig.savefig('map_ancient.png')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv("samples_dates.txt", delimiter="\t")

# Generate random y values between 0 and 1 for demonstration
data['y'] = np.random.rand(len(data))

# Create a color map for different methods
color_map = {
    'radiocarbon': 'blue',
    'context': 'green'
}

# Map the method categories to colors
data['color'] = data['method'].map(color_map)

# Sort the data by age in descending order
data = data.sort_values(by='age', ascending=False)

# Scatter plot
plt.figure(figsize=(10, 2))
plt.scatter(data['age'], data['y'], c=data['color'], alpha=0.6)

# Hide Y-axis ticks and labels
plt.yticks([])

# Add labels and title
plt.xlabel('Age (years cal BP)')

# Set x-axis ticks at intervals of 1000 and rotate x-axis labels
plt.xticks(rotation=45)
plt.xticks(np.arange(0, data['age'].max()+1000, 1000))

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv("samples_dates.txt", delimiter="\t")

# Generate random y values between 0 and 1 for demonstration
data['y'] = np.random.rand(len(data))

# Create a color map for different origins
color_map = {
    'internal': '#e3aa4d',
    'external': '#41534f'
}

# Create a marker map for different methods
marker_map = {
    'radiocarbon': 'o',  # Circle
    'context': 'o'  # Square
}

# Map the origin categories to colors
data['color'] = data['origin'].map(color_map)

# Map the method categories to markers
data['marker'] = data['method'].map(marker_map)

# Sort the data by age in ascending order
data = data.sort_values(by='age')

# Scatter plot
plt.figure(figsize=(10, 2))
for index, row in data.iterrows():
    edgecolor = 'black' if row['method'] == 'radiocarbon' else None
    markersize = 40 if row['method'] == 'radiocarbon' else 25
    plt.scatter(row['age'], row['y'], c=row['color'], marker=row['marker'], edgecolor=edgecolor, linewidth=1, alpha=0.8, s=markersize)

# Add labels and title
plt.xlabel('Age (years cal BP)')

# Set x-axis ticks at intervals of 1000 and rotate x-axis labels
plt.xticks(rotation=45)
plt.xticks(np.arange(0, data['age'].max()+1000, 1000))

# Hide Y-axis ticks and labels
plt.yticks([])

# Show plot
plt.tight_layout()
plt.savefig('dates.png')

plt.show()
```
